```python
# Copy original dataframe
df_fe = df_mat.copy()

# Remove cheating columns
df_fe = df_fe.drop(columns=["G1", "G2"])

# New engineered features
df_fe["absence_rate"] = df_fe["absences"] / (df_fe["age"] * 10)
df_fe["failure_ratio"] = df_fe["failures"] / (df_fe["age"] + 1)
df_fe["study_load"] = df_fe["studytime"] * df_fe["traveltime"]
df_fe["social_score"] = df_fe["goout"] + df_fe["freetime"] + df_fe["Walc"]
df_fe["family_support"] = df_fe["famrel"] + df_fe["Medu"] + df_fe["Fedu"]
```

```python
X = df_fe.drop("G3", axis=1)
y = df_fe["G3"]

X_encoded = pd.get_dummies(X, drop_first=True)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest RMSE:", rmse_rf)
print("Random Forest R2:", r2_rf)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_encoded)

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_s.shape[1],)),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(1)
])

model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

model.summary()
```

```python
history = model.fit(
    X_train_s, y_train_s,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    verbose=0
)
```

```python
y_pred_nn = model.predict(X_test_s).flatten()

mse_nn = mean_squared_error(y_test_s, y_pred_nn)
rmse_nn = np.sqrt(mse_nn)
r2_nn = r2_score(y_test_s, y_pred_nn)

print("Neural Net RMSE:", rmse_nn)
print("Neural Net R2:", r2_nn)
```
